```python
from google.colab import files

mp3_audio = AudioSegment.from_file("baacaguaaxkbaaj86mhevsupsy4vqwic9r-sljckhwroaaiagqacqwd5vt7zwp47-vocals_ysBu9UV0.mp3", format="mp3")
mp3_audio.export("audio.wav", format="wav")
```

```python
audio, sr = librosa.load("audio.wav")

plt.figure(figsize=(12, 3))
librosa.display.waveshow(audio, sr=sr)
plt.title("Gelombang Suara - Original")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

```python
D = np.abs(librosa.stft(audio))
DB = librosa.amplitude_to_db(D, ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(DB, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
plt.title('Spektrogram Frekuensi - Original')
plt.tight_layout()
plt.show()
```

```python
noise = np.random.normal(0, 0.02, audio.shape)
audio_noise = audio + noise

plt.figure(figsize=(12, 3))
librosa.display.waveshow(audio_noise, sr=sr)
plt.title("Dengan Noise")
plt.tight_layout()
plt.show()
```

```python
amplified = audio * 1.5

plt.figure(figsize=(12, 3))
librosa.display.waveshow(amplified, sr=sr)
plt.title("Amplify (Volume Ditingkatkan)")
plt.tight_layout()
plt.show()
```

```python
sf.write('hasil_diproses.wav', amplified, sr)

files.download('hasil_diproses.wav')
```
